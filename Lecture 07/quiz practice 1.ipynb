```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 1000
vec_x = np.random.normal(loc = 2, scale = 5, size = n)
vec_y = np.random.normal(loc = 2, scale = 5, size = n)

fig, list = plt.subplots(1,2,figsize=(10,15))
plt.tight_layout


list[0].hist(x = vec_x)
list[0].set_title("Random sample from 1000")
list[0].set_xlabel("Value")
list[0].set_ylabel("Frequency")

list[1].hist(x = vec_y)
list[1].set_title("Second Random sample from 1000")
list[1].set_xlabel("Value")
list[1].set_ylabel("Frequency")
```

```python
vec_x = np.random.normal(loc = 2, scale = 5, size = 1000)

vec_y = (vec_x) ** 2 + 2 * (vec_x) + 5

plt.scatter(vec_x, vec_y, color = 'blue')
plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("y against x")
plt.show
```

```python

csv_files = ['data/wdi_2000.csv','data/wdi_2020.csv','features.csv']

for i, csv_file in enumerate(csv_files):
    df = pd.read_csv(csv_file)
    transformed_variable = (df['var1'] ** 2)
    plt.hist(transformed_variable)

    plt.figtext(0.9, 0.05, f'Figure {i+1}', fontsize=12, ha='right')
    plt.xlabel('Transformed Variable')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Transformed Variable for {csv_file}')
    plt.show()
```
