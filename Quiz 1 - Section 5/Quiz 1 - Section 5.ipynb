(a) Create two random variables ...

- Generate "vec_x" from a uniform distribution with (low = 10, high = 20, size = 1000)
- Generate "vec_z" from a uniform distribution with (low = -2, high = 5, size = 1000)
- Plot a histogram of each variable
    - Label the axes and title
    - Each graph should have a different title
 
To get full points, plot the histograms on a grid (See Lecture 6)

```python
# Write your own code


n = 1000
vec_x  = np.random.uniform(low = 10,high = 20, size = n)
vec_z  = np.random.uniform(low = -2,high = 5, size = n)

fig, list = plt.subplots(1,2,figsize=(8,5))
plt.tight_layout


list[0].hist(x = vec_x)
list[0].set_title("Random sample from 1000")
list[0].set_xlabel("Value")
list[0].set_ylabel("Frequency")

list[1].hist(x = vec_z)
list[1].set_title("Second Random sample from 1000")
list[1].set_xlabel("Value")
list[1].set_ylabel("Frequency")

```
